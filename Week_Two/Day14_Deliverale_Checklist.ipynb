##Problem Statement
Financial institutions face significant risk from fraudulent or anomalous transactions. Detecting these anomalies early is critical to reducing financial loss and improving trust.

This project investigates the use of machine learning techniques to identify anomalous financial transactions based on transaction amount and time-related features.

##Dataset Description
The dataset consists of financial transaction records containing numerical features such as transaction amount and time-based variables. The data was preprocessed by handling missing values, scaling numerical features, and ensuring compatibility with machine learning algorithms.

##Methodology
An unsupervised learning approach was adopted due to the absence of labeled anomaly data. The Isolation Forest algorithm was selected because it is effective in detecting rare and extreme observations in high-dimensional financial datasets.

Features were standardized to ensure equal contribution during model training.

After evaluating Logistic Regression, Random Forest, and XGBoost, XGBoost was selected as the final model due to its superior ROC-AUC, robustness to class imbalance, and ability to model complex non-linear relationships, while still maintaining interpretability through SHAP-based explanations.

##Results
The model achieved an accuracy of 81%. However, further analysis using the confusion matrix revealed a strong ability to correctly classify normal transactions, while recall for anomalous transactions was comparatively low. This indicates that while the model is conservative, it may miss certain rare anomalies.
